```python
import math

C = 40
H = 20

input_str = input("Enter comma-separated values for D: ")  # "100,150,180"
D_values = input_str.split(',')

results = []
for d in D_values:
    D = int(d)
    Q = math.sqrt((4 * C * D) / H)
    results.append(str(round(Q)))

print(",".join(results))
```

```python
X = int(input("Enter number of rows (X): "))
Y = int(input("Enter number of columns (Y): "))
result = []

for i in range(X):
    row = []
    
    for j in range(Y):
        value = (i ** 2) + (j ** 2)
        row.append(value)
    result.append(row)

print("Generated 2D list:")
for row in result:
    print(row)
```

```python
binary_input = input("Enter comma-separated 4-digit binary numbers: ")

binary_numbers = binary_input.split(',')
result = []

for binary in binary_numbers:
    decimal = int(binary, 2)
    if decimal % 7 == 0:
        result.append(binary)

print("Binary numbers divisible by 7:")
print(','.join(result))
```

```python
balance = 0
print("Enter transactions (D/W amount): ")

while True:
    transaction = input()
    
    if transaction.lower() == '':
        break
    parts = transaction.split()
    type, amount = parts[0], parts[1]
    amount = int(amount)

    if type.upper() == 'D':
        balance += amount
    elif type.upper() == 'W':
        balance -= amount
    else:
        print("Invalid transaction type. Use 'D' for deposit or 'W' for withdrawal.")

print("Final bank balance:", balance)
```
