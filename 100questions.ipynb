```python
num = int(input('enter a number: '))

if num > 0:
    print('positive')
elif num < 0:
    print('negitive')
else:
    print('zero')
```

```python
num = int(input('enter a number: '))

if num % 2 == 0:
    print('even')
else:
    print('odd')
```

```python
num = int(input('enter a number: '))

if num % 4 == 0:
    print('yes, its a leap year')
else:
    print('nope, its isnt a leap year')
```

```python
ch = input('enter a text: ')

if ch.isalpha():
    print("yes, its a alphabet")
else:
    print("nope, its not a alphabet")
```

```python
alphabet = input('enter a alphabet: ')

vowels = set('aeiou')

if alphabet in vowels:
    print('its a vowel')
else:
    print('its a consonant')
```

```python
ch = input('enter a text: ')

if ch.isalpha():
    print("its a alphabet")
elif ch.isnumeric():
    print("its a digit")
else:
    print("its a special char")
```
