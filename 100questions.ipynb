```python
amount = int(input('enter total amount: '))

notes = [500, 200, 100, 50, 20, 10, 5, 2, 1]

total_notes = 0

for note in notes:
    if amount >= note:
        count = amount // note
        amount = amount % note
        total_notes += count

print(total_notes)
```

```python
a1 = int(input('enter angle 1 of triangle: '))
a2 = int(input('enter angle 2 of triangle: '))
a3 = int(input('enter angle 3 of triangle: '))

total = a1 + a2 + a3

if a1 > 0 and a2 > 0 and a3 > 0 and total == 180:
    print('yes, this is a valid triangle!')
    if a1 == a2 == a3:
        print('its an equilateral triangle')
    elif (a1 == a2) or (a2 == a3) or (a3 == a1):
        print('its an isosceles triangle')
    else:
        print('its a scalene triangle')
else:
    print('no, this is not a valid triangle!')
```

```python
num = int(input("Enter a number: "))
rev = 0

while num != 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed number:", rev)
```

```python
num = int(input("Enter a number: "))

number = num
rev = 0

while num != 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print(number == rev)
```

```python
num = int(input('Enter a number: '))

for i in range(2, num + 1):
    is_prime = True
    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i, "is a prime number")
```

```python
num = int(input("Enter a number: "))

num_digits = len(str(num))

temp = num
sum_of_powers = 0

while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** num_digits
    temp = temp // 10

if sum_of_powers == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is NOT an Armstrong number")
```
