```python
num = int(input("Enter a number: "))

num_digits = len(str(num))

temp = num
sum_of_powers = 0

while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** num_digits
    temp = temp // 10

if sum_of_powers == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is NOT an Armstrong number")
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def is_armstrong(n):
    num_digits = len(str(n))
    temp = n
    sum_of_powers = 0
    while temp > 0:
        digit = temp % 10
        sum_of_powers += digit ** num_digits
        temp //= 10
    return sum_of_powers == n

num = int(input("Enter a number: "))

if is_prime(num):
    print(num, "is a Prime number")
else:
    print(num, "is NOT a Prime number")

if is_armstrong(num):
    print(num, "is an Armstrong number")
else:
    print(num, "is NOT an Armstrong number")
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
found = False

for i in range(2, num // 2 + 1):
    if is_prime(i) and is_prime(num - i):
        print(f"{num} = {i} + {num - i}")
        found = True

if not found:
    print(num, "cannot be expressed as the sum of two prime numbers")
```

```python
def reverse_sentence(sentence):
    words = sentence.split()

    def helper(index):
        if index == len(words) - 1:
            return words[index]
        return helper(index + 1) + " " + words[index]

    return helper(0)

# Example usage
sentence = input("Enter a sentence: ")
reversed_sentence = reverse_sentence(sentence)
print("Reversed sentence:", reversed_sentence)
```

```python

n = int(input('enter number of elements: '))
arr = []

for i in range(n):
    value = int(input(f'enter the number {i+1}: '))
    arr.append(value)

def calculate_avg(nums):
    total = 0
    for num in nums:
        total += num
    average = total//len(nums)
    return average

avg = calculate_avg(arr)
print(avg)
```

```python
n = int(input('enter number of elements: '))
arr = []

for i in range(n):
    value = int(input(f'enter the number {i+1}: '))
    arr.append(value)

def find_largest(nums):
    largest = nums[0]
    for num in nums:
        if num > largest:
            largest = num
    return largest

large = find_largest(arr)
print(large)
```
