```python
s = 'abc12def34gh5'

# find the sum of the all intergers
temp = ''
total = 0

for i in s:
    if i.isdigit():
        temp += i
    else:
        if temp != '':
            total += int(temp)
            temp = ''
if temp != '':
    total += int(temp)
print(total)
            
```

```python
s = "((()))"

count = 0
for i in s:
    if i == '(':
        count += 1
    elif i == ')':
        count -= 1
    if count < 0:
        print('imbalanced')
        break
else:
    if count == 0:
        print('balanced')
    else:
        print('imbalanced')
```

```python
sentence = 'cat dog dog bird cat cat'

arr = sentence.split(' ')
word_count = {}

for i in arr:
    word_count[i] = word_count.get(i,0) + 1
most = max(word_count, key=word_count.get)
print(most)
```
