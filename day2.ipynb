```python
s = 'abc12def34gh5'

# find the sum of the all intergers
temp = ''
total = 0

for i in s:
    if i.isdigit():
        temp += i
    else:
        if temp != '':
            total += int(temp)
            temp = ''
if temp != '':
    total += int(temp)
print(total)
            
```

```python
s = "((()))"

count = 0
for i in s:
    if i == '(':
        count += 1
    elif i == ')':
        count -= 1
    if count < 0:
        print('imbalanced')
        break
else:
    if count == 0:
        print('balanced')
    else:
        print('imbalanced')
```

```python
sentence = 'cat dog dog bird cat cat'

arr = sentence.split(' ')
word_count = {}

for i in arr:
    word_count[i] = word_count.get(i,0) + 1
most = max(word_count, key=word_count.get)
print(most)
```

```python
s = 'crypt'

vowels = set('aeiou')

def vowel_check(s):
    for i in range(len(s)):
        if s[i] in vowels:
            return i
    return -1

first_index = vowel_check(s)
if first_index == -1:
    print(f'no vowels found!')
else:
    print(f'first vowel index {first_index}')
        
```

```python
s = 'education'

arr = list(s)
def moving_vowels(arr):
    vowels = set('aeiou')
    for i in arr:
        if i in vowels:
            arr.remove(i)
            arr.append(i)
    arr = ''.join(arr)
    return arr

print(moving_vowels(arr))
```

```python
s = 'hello world'
vowels = set('aeiou')

for i in s:
    if i in vowels:
        s = s.replace(i, i.upper())
print(s)
```
