```python
word = 'vishwa'
vowels = ['a','e','i','o','u']

def count_vowels(text):
    count = 0
    for i in text:
        if i in vowels:
            count  += 1
    return count

print(count_vowels(word))
```

```python
text = 'Service Based Companies'

def case_counter(text):
    upper_count = 0
    lower_count = 0
    for i in text:
        if i.isupper():
            upper_count += 1
        elif i.islower():
            lower_count += 1
    return upper_count, lower_count

upper, lower = case_counter(text)
print(f'upper case letters: {upper}, lower case letters: {lower}')
```

```python
arr = [1,2,3,2,3,1,5,6,7]

def finding_duplicates(arr):
    new = set()
    duplicates = set()
    for i in arr:
        if i in new:
            duplicates.add(i)
        else:
            new.add(i)
    return new, duplicates

new, duplicates = finding_duplicates(arr)
print(duplicates)
```

```python
nums = [2,12,4,12,1,3]

def sum_n_multiple(nums):
    result = 1
    if len(nums) % 2 == 0:
        for i, j in enumerate(nums):
            if i%2 == 0:
                sum = j+nums[i+1]
                result *= sum
            else:
                continue
    return result

print(sum_n_multiple(nums))
```

```python
nums = [2,12,4,12,1,3]

def sum_n_multiple(nums):
    result = 1
    for i in range(0, len(nums), 2):
        sum = nums[i] + nums[i+1]
        result *= sum
    return result

print(sum_n_multiple(nums))
```
